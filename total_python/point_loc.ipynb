```python
import folium
import pandas as pd

# define the world map
world_map = folium.Map()

# display world map
#world_map
```

```python
# centre of Dundee latitude and longitude values
latitude = 56.4587397
longitude = -2.9820049

Center_loc = folium.map.FeatureGroup()
Center_loc.add_child(
    folium.CircleMarker(
        [latitude, longitude],
        radius=7, 
        color='black',
        fill=True,
        fill_color='purple',
        fill_opacity=0.4
    )
)
# Create map and display it
center_map = folium.Map(location=[latitude, longitude], zoom_start=24)

# Display the map of centre of Dundee
center_map.add_child(Center_loc)
```

# Interest point

```python
# Read Dataset 
Idata = pd.read_excel('old_data/Interest_points.xlsx')
Idata = Idata.rename(columns={'grid number':'grid_number'})
Idata.head()
```

```python
Interest_loc = folium.map.FeatureGroup()

for lat, lng, in zip(Idata.Latitude, Idata.Longitude):
    Interest_loc.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, 
            color='grey',
            fill=True,
            fill_color='purple',
            fill_opacity=0.4
        )
    )

# Add Interest point to map
Iloc_map = folium.Map(location=[latitude, longitude], zoom_start=12)
Iloc_map.add_child(Interest_loc)
```

# current point

```python
cdata = pd.read_excel('old_data/Project_data/Charging_points.xlsx')
cdata = cdata.rename(columns={'grid number':'grid_number'})
cdata = cdata.rename(columns={'CP ID':'CP_ID'})
cdata.head()
```

```python
Current_loc = folium.map.FeatureGroup()

for lat, lng, in zip(cdata.Latitude, cdata.Longitude):
    Current_loc.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, 
            color='grey',
            fill=True,
            fill_color='purple',
            fill_opacity=0.4
        )
    )

cloc_map = folium.Map(location=[latitude, longitude], zoom_start=12)
cloc_map.add_child(Current_loc)
```

```python
latitudes = list(cdata.Latitude)
longitudes = list(cdata.Longitude)
labels = list(cdata.grid_number)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(cloc_map)    
    
cloc_map.add_child(Current_loc)
```
